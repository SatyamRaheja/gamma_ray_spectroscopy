### Preamplifier
The major focus will be on High Purity Germanium (HPGe) Detectors. The analytical form of pre-amplifier signal for High Purity Germanium (HPGe) detector is given by Jordanov and co-workers as follows:-
$$V(t)=-\frac{1}{\tau-\theta}\left(e^{\left(-\dfrac{t}{\tau}\right)}-e^{\left(-\dfrac{t}{\theta}\right)}\right)$$
where $\theta$ is finite rise time and $\tau$ is decay time 

The polarity of preamp output signal is generally negative if current pulse will go into the preamp input having negative feedback circuit.

```python
# Importing libraries
import numpy as np # for mathematical functions and arrays
import matplotlib.pyplot as plt # for plotting
```

```python
# Defining function for pre-amplifier signal for HPGe
def pre_amplifier_output_HPGe(t,tau,theta): # t,tau and theta are parameters of function that we can change while calling the function
    # t is time array, tau is decay time and theta is rise time
    V=-(np.exp(-t/tau)-np.exp(-t/theta))/(tau-theta) # analytical formula of pre-amplifier signal for HPGe detector
    return V
```

```python
t=np.linspace(0,50,100) 
# t is time array, 1st and 2nd parameter is the interval divided in number of steps(equally spaced) given as 3rd parameter 
# Let's plot the preamp output voltage as function of time with the varying parameters (tau,theta)
tau_11=2 # In variable name, first number for row and second number for column
theta_11=1
tau_12=5
theta_12=1
tau_13=10
theta_13=1
tau_21=10
theta_21=2
tau_22=10
theta_22=5
tau_23=10
theta_23=8
# Displayed them as subplots to oberve the variation easily
fig, ax = plt.subplots(2, 3) #to subplot in 2 rows and 3 columns, all subplots are sharing x axis
fig.set_size_inches(18,10) # size of figure
fig.tight_layout(pad=6) # spacing between subplots

# calling pre_amplifier_output_HPGe function
V1_preamp=pre_amplifier_output_HPGe(t,tau_11,theta_11) 
V2_preamp=pre_amplifier_output_HPGe(t,tau_12,theta_12) 
V3_preamp=pre_amplifier_output_HPGe(t,tau_13,theta_13) 
V4_preamp=pre_amplifier_output_HPGe(t,tau_21,theta_21) 
V5_preamp=pre_amplifier_output_HPGe(t,tau_22,theta_22) 
V6_preamp=pre_amplifier_output_HPGe(t,tau_23,theta_23) 
# making array of diferent output voltages
V_preamp=np.array([[V1_preamp,V2_preamp,V3_preamp],[V4_preamp,V5_preamp,V6_preamp]])
# adding super title and individual titles for subplots
fig.suptitle("Pre-amplifier signal for HPGe Detector",fontsize=16)
# setting the title
ax[0, 0].set_title(r'Parameters: $\tau=2$, $\theta=1$',fontsize=16)
ax[0, 1].set_title(r'Parameters: $\tau=5$, $\theta=1$',fontsize=16)
ax[0, 2].set_title(r'Parameters: $\tau=10$, $\theta=1$',fontsize=16)
ax[1, 0].set_title(r'Parameters: $\tau=10$, $\theta=2$',fontsize=16)
ax[1, 1].set_title(r'Parameters: $\tau=10$, $\theta=5$',fontsize=16)
ax[1, 2].set_title(r'Parameters: $\tau=10$, $\theta=8$',fontsize=16)

# plotting vertical line passing from minima (for better comparison)

# Loop to run command in all subplots
for i in range(0,2):
    for j in range(0,3):
        ax[i][j].xaxis.set_tick_params(labelsize=16)
        ax[i][j].yaxis.set_tick_params(labelsize=16)
        ax[i][j].plot(t,V_preamp[i][j]) # plotting output voltage vs time
        # Next 4 lines are to plot a vertical line corresponding to minimum voltage
        min_V_index=np.where(V_preamp[i][j]==min(V_preamp[i][j]))[0][0]
        y=np.linspace(0,min(V_preamp[i][j]),10)
        x=np.array([t[min_V_index]]*10)
        ax[i][j].plot(x,y,linestyle='dashed')
        ax[i][j].set_xlabel("Time",fontsize=16) # labelling the axes
        ax[i][j].set_ylabel("Voltage",fontsize=16)
        ax[i][j].grid() # adding gridlines
plt.show()
```

#### Oberve the following points:-

1) In above plots, we can easily observe the variation of decay time row-wise and the variation of rise time column-wise.

2) As the decay time increases, the curve is getting more broader and taking more time to decay.

3) While as the rise time increase, the curve minima is shifting to right side i.e. signal is taking more time to rise.

#### For HPGe detectors, the rise time of the preamplifier signal is in few nanoseconds while the decay time is in hundreds of microseconds. Thus, its output signal can be approximated as a step pulse.
